```python
import ollama

# chat with model
result = ollama.chat(
    model='deepseek-r1:1.5b',
    messages=[{
        'role': 'user',
        'content': 'What are the steps involved in baking a chocolate cake?'
    }],
)

print(result.message.content)
```
